# Introduction/ Business Problem 

This report examines the venues that are present around The Peel Region area in Ontario, Canada using the Foursquare API to collect data for these venues. The goal of this report is to create an algorithm that will give location regions where one should open their venue given the category of venue it is. This will be done using Foursquare's search and explore tools which allow us to see the amount of venues with a similar category, and observe the trends of that area using the trending command on the API. The Peel Region consists of 3 cities, Caledon, Brampton, and Mississauga and data for this region will be collected off of the internet. Imagine an entrepreneur wants to open a Chinese grocery store but doesn't know which region to open such store. This algorithm will give the entrepreneur some options on which locations are best for their decision based on the ethnicity of the general population in a given city and the amount of venues with the same business idea in the area already. The higher the population that will be in favour for the store opening, the better "score" it gets regarding the final decision made, but if there is already multiple stores in the area with the same business idea,  the lower "score" it will receive. This is just a simple example and there will be many other factors that go into the final decision by the algorithm.
